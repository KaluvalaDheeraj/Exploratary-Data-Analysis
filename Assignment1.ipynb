```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plot
import numpy as np
```

```python
haberman=pd.read_csv('haberman.csv')
```

```python
print(haberman.columns)  #The names of Columns presented in the Data set
```

```python
haberman.plot(kind='scatter', x='survival', y='age') ;   #2D Scatter Plot of Age and Survial
plot.show()
```

```python
haberman.plot(kind='scatter',x='age',y='survival')
plot.show()
```

```python
haberman.plot(kind='scatter',x='year',y='age')
plot.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(haberman, hue="survival", size=4) \
   .map(plot.scatter, "axillary", "age") \
   .add_legend();
plot.show();
#Note : Here, The Orange and Blue points are mixed one in one and the Orange points are easily identified.. 
#It can't be seperated easily
```

```python
plot.close();
sns.set_style("whitegrid");
sns.pairplot(haberman,hue="survival",x_vars=["age","year", "axillary"],y_vars=["age","year", "axillary"]);
plot.show()
```

#  HISTOGRAM,PDF,CDF

```python
sns.FacetGrid(haberman, hue="survival", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plot.show();
```

```python
sns.FacetGrid(haberman, hue="survival", size=5) \
   .map(sns.distplot, "year") \
   .add_legend();
plot.show();
```

```python
sns.FacetGrid(haberman, hue="survival", size=5) \
   .map(sns.distplot, "axillary") \
   .add_legend();
plot.show();
```

```python
import numpy as np
haberman_1 = haberman.loc[haberman["survival"] == 1];
haberman_2 = haberman.loc[haberman["survival"] == 2];

plot.plot(haberman_1["year"], np.zeros_like(haberman_1 ['survival']), 'o')
plot.plot(haberman_2["year"], np.zeros_like(haberman_2 ['survival']), 'o')

plot.show()
```

Mean, median Standard deviation

```python
#Mean, Variance, Std-deviation,  
print("Means:")
print(np.mean(haberman_1["year"]))
#Mean with an outlier.
print(np.mean(np.append(haberman_1["year"],60)));
print(np.mean(haberman_2["year"]))

print("\nStd-dev:");
print(np.std(haberman_1["year"]))
print(np.std(haberman_2["year"]))
```
